```python
colors = {
    'yellow': [255, 255, 0],  #yellow
    'black': [0, 0, 0],#black
    'green': [0,255,0],
    'blue': [0,0,255],
    'brown':[181,101,29]
   }
```

```python
path='C:/Users/sriram/Downloads/SIH/Clipped_NDVI/'
f=os.listdir(path)
img=cv2.imread(path+f[6],0)
print(img.shape)
```

```python
imgtotry=[f[0],f[6],f[13]]

for file in imgtotry:
    img=cv2.imread(path+file,0)
    new_img=np.zeros((len(img),len(img[0]),3))
    y=0
    g=0
    b=0
    for i in range(len(img)):
        for j in range(len(img[0])):
            if img[i][j]<50:
                new_img[i][j]=colors['yellow']
                y+=1
            elif img[i][j]<=100:
                new_img[i][j]=colors['green']
                g+=1
            else:
                new_img[i][j]=colors['black']
                b+=1
    print('\n\nyellow:',y,'\ngreen',g,'\nblack',b) 
    plt.imsave(file.split('.')[0]+'.png',new_img)
    
```
